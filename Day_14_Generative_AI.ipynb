```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = "The hero enters the ancient forest and discovers"

result = generator(
    prompt,
    max_length=80,      # maximum length of generated text
    num_return_sequences=1,  # number of generated outputs
    temperature=0.8     # randomness (0.7–1.0 is good for creativity)
)

# Print generated story
print(result[0]["generated_text"])
```

```python
from transformers import BertTokenizerFast

tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

patient_query = "I have chest pain and shortness of breath"

encoded = tokenizer(
    patient_query,
    padding="max_length",
    truncation=True,
    max_length=20,
    return_tensors="pt"
)

tokens = tokenizer.convert_ids_to_tokens(encoded["input_ids"][0])

print("Patient Query:", patient_query)
print("Tokens:", tokens)
print("Token IDs:", encoded["input_ids"][0].tolist())
print("Attention Mask:", encoded["attention_mask"][0].tolist())
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London. Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024."
doc = nlp(text)

entities = {
    "Teams": [],
    "Players": [],
    "Locations": [],
    "Dates": [],
    "Numbers": []
}


for ent in doc.ents:
    if ent.label_ == "ORG":
        entities["Teams"].append(ent.text)
    elif ent.label_ == "PERSON":
        entities["Players"].append(ent.text)
    elif ent.label_ in ["GPE", "LOC", "FAC"]:
        entities["Locations"].append(ent.text)
    elif ent.label_ == "DATE":
        entities["Dates"].append(ent.text)
    elif ent.label_ == "CARDINAL":
        entities["Numbers"].append(ent.text)

print("Structured Insights from Match Report:\n")
for key, value in entities.items():
    print(f"{key}: {value}")
```

```python
!pip install gensim nltk

import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

nltk.download('punkt')
nltk.download('punkt_tab')

corpus = [
    "Smart manufacturing uses sensors and telemetry to monitor machines",
    "Predictive maintenance reduces machine downtime",
    "Quality control ensures products meet standards",
    "Supply chain management coordinates production and delivery",
    "Telemetry data helps detect faults in manufacturing equipment",
    "Maintenance teams repair machines and improve efficiency",
    "Quality assurance improves product reliability"
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]


model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=3,
    min_count=1,
    workers=2
)

print("Words similar to 'maintenance':")
print(model.wv.most_similar("maintenance"))

print("\nWords similar to 'quality':")
print(model.wv.most_similar("quality"))

print("\nSimilarity between 'maintenance' and 'machines':")
print(model.wv.similarity("maintenance", "machines"))
```

```python

!pip install gensim nltk

import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec


nltk.download('punkt')
nltk.download('punkt_tab')


corpus = [
    "Doctors diagnose diseases and recommend treatment for patients",
    "Patients report symptoms to doctors for proper diagnosis",
    "Medication helps treat diseases and improve patient health",
    "Hospitals provide treatment and care for patients",
    "Symptoms help doctors identify diseases",
    "Medical research improves diagnosis and treatment methods",
    "Doctors prescribe medication to patients"
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=3,
    min_count=1,
    workers=2
)

print("Words similar to 'treatment':")
print(model.wv.most_similar("treatment"))

print("\nWords similar to 'patients':")
print(model.wv.most_similar("patients"))

print("\nSimilarity between 'diagnosis' and 'symptoms':")
print(model.wv.similarity("diagnosis", "symptoms"))
```
